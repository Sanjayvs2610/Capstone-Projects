### Task 2 - Identify the null values and perform null values treatment

```python
df.isnull().sum()
```

##### There are some null values present in the dataset

##### Cleaning the null values from the columns
- For categorical we have to replace it by using mode of the column 
- For numerical column we have to treat it with the mean or median of the columns 

```python
numeric_columns = []
categorical_columns = []

for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        numeric_columns.append(column)
    else:
        categorical_columns.append(column)

table = [numeric_columns, categorical_columns]
print(tabulate({"numerical":numeric_columns,
                "categorical": categorical_columns}, headers = ["numerical", "categorical"]))
```

```python
def info_of_cat(col):
    print(f"Unique values in {col} are :{df[col].unique()}")
    print(f"Mode of {col} is : {df[col].mode()[0]}")
    print(f"Number of missing values in {col} is : {df[col].isnull().sum()}")
```

#### Performing null values treatment for the Homepage column

```python
info_of_cat("homepage")
```

```python
df["homepage"].fillna(" http://www.missionimpossible.com/",inplace=True)
```

#### Overview column

```python
info_of_cat("overview")
```

```python
df["overview"].fillna("'In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission,but becomes torn between following orders and protecting an alien civilization.'",inplace=True)
```

#### Release date column

```python
info_of_cat("release_date")
```

```python
df["release_date"].fillna("01-01-2006",inplace=True)
```

#### Tagline column

```python
info_of_cat("tagline")
```

```python
df["tagline"].fillna("Based on a true story",inplace=True)
```

#### Performing null tretament for the numerical columns  

```python
a = df["runtime"].mean()
a
```

```python
df["runtime"].fillna(a,inplace=True)
```

```python
df.isnull().sum()
```

##### All the null values presentin the data has been replaced , Now there is 0 null values in the dataset
